```python
# import statement 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
```

```python
# question 1 
plt.figure(figsize = (10,6))
sns.countplot(y='state', data=df)
plt.title('Company count by state')
plt.show()
```

```python
# question 3
sns.barplot(x='Industry', y='job_openings', data=df, estimator = 'sum', errorbar = None)
plt.xticks(rotation = 90)
plt.title('Jobs open in Industry')
plt.show()
```

```python
# question 5
df_rate = df.groupby('founded')['growth_percentage'].mean().reset_index()
sns.lineplot(x='founded', y='growth_percentage', data=df_rate)
plt.title('Average Growth by Year Founded')
plt.show()
```

```python
# question 7
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.title('Employee Estimated Revenues in millions')
plt.show()
```

```python
# question 9 
sns.histplot(x='founded', data=df)
plt.title('Numbers of Companies Founded each year')
plt.show()
```
