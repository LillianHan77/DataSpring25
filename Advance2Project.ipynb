```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
# Create a scatter plot that shows land size on the x-axis and price on the y-axis. 
# Change the color of the points to reflect the type of the property. Paste the image below.
sns.scatterplot(data=df, x= 'Landsize', y= 'Price', hue= 'Type', palette='Set1')
plt.title('Landsixe vs. Price by Property Type')
plt.show()
```

```python
#Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis. 
#You will need to use aggregation and group by to accomplish this. Paste the image below.
avg_Price= df.groupby('YearBuilt')['Price'].mean().reset_index()
sns.lineplot(x= 'YearBuilt', y= 'Price', data= avg_Price)
plt.title('Average Price Per Year Built')
plt.show()
```

```python
#Create a bar chart that shows the region on the x-axis and the average price of the property on the y-axis. 
#You will need to use aggregation and group by to accomplish this. Paste the image below.
PriceAvg= df.groupby('Regionname')['Price'].mean().reset_index()
sns.barplot(x= 'Regionname', y= 'Price', data= PriceAvg)
plt.title('Average Price by Region')
plt.xticks(rotation=90)
plt.show()
```

```python
#Create a histogram of property prices. Paste the image below.
sns.histplot(x= 'Price', data=df)
plt.show()
```

```python
#Create a box plot that shows the years that each property was built on the x-axis and the region on the y-axis. 
#These box plots should be oriented horizontally. Paste an image of the graphic below.
sns.boxplot(x= 'YearBuilt', y= 'Regionname', data=df, orient='h')
plt.show()
```
