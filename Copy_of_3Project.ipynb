```python
import pandas as pd
df = pd.read_csv('/content/data_analytics_salaries.csv')
```

```python
# rowa then columns
df.shape
```

```python
# average salary
round(df['salary_in_usd'].mean(), 2)
```

```python
# average salary for full time
avg_ft = round(df[df['employment_type'] == 'FT']['salary_in_usd'].mean(), 2)
print(avg_ft)
```

```python
# highest remote work
df.groupby('experience_level')['remote_ratio'].mean().sort_values(ascending= False)
```

```python
# avg remote ratio to 2 decimals
round(df.groupby('experience_level')['remote_ratio'].mean(), 2)
```

```python
# rounded to the nerest tenth
round(df.groupby('experience_level')['remote_ratio'].mean(), 1)
```

```python
# increase of remote working in 2022
avg_2021 = df[df['work_year'] == 2021].groupby('experience_level')['remote_ratio'].mean()
avg_2022 = df[df['work_year'] == 2022].groupby('experience_level')['remote_ratio'].mean()

increase = (avg_2022 - avg_2021).round(1)

biggest_increase = increase.sort_values(ascending=False)

print(biggest_increase)
```

```python
# jobs in US
df[df['company_location']=='US']['job_title'].value_counts(ascending= True)
```

```python
# medain salary of data analyst
df[df['job_title']== 'Data Analyst']['salary_in_usd'].median()
```

```python
# standard deviation salary for machine learning engineer
df[df['job_title']== 'Machine Learning Engineer']['salary_in_usd'].std()
```

```python
# avg salary for data engineer
round(df[df['job_title']== 'Data Engineer']['salary_in_usd'].mean(), 1)
```

```python
# avg salary of data analysts years through 2020 and 2022
data_analytics = df[df['job_title'] == 'Data Analyst']
avg_salaries = data_analytics.groupby('work_year')['salary_in_usd'].mean().round(2)
avg_filtered_sal = avg_salaries.loc[[2020, 2022]]
print(avg_filtered_sal)
```

```python
# avg salary in 2021 for data analysts
job_analyst = df[df['job_title'] == 'Data Analyst']
salary_avg = job_analyst.groupby('work_year')['salary_in_usd'].mean().round(1)
average_2021 = salary_avg.loc[2021]
print(average_2021)
```
